```python
%matplotlib inline
import numpy as np
import math as mt
import matplotlib
import matplotlib.pyplot as plt
from pylab import *
from scipy.stats import *
```

```python
data = np.loadtxt('6.csv', delimiter=',', skiprows=3)
print(len(data))
```

```python
def BayesEst(x, alpha, beta):
    return (alpha + len(x))/(beta + sum(x))
```

```python
lmd = 1./66.
t_0 = 200
t = 70000
params = [(1,1),(20,1),(1,30),(120,2),(2,120),(60,1)]

# Сюда запишу время между i-м моментом выхода из строя сервера и (i+1)-м
delta = []
for i in range(len(data)-1):
    delta.append(data[i+1] - data[i])

# plot(delta)
# hist(delta)
# print mean(delta)

est = np.zeros((len(params), len(data)))
for i in range(len(params)):
    est[i] = np.array([BayesEst(delta[:(j+1)], params[i][0], params[i][1]) for j in range(len(data))])

figure(figsize=(12,5))
ylim((0,0.01))
xlabel("Sample size")
for i in range(len(params)):
    plot(abs(lmd-est[i]), label="Parameters of Prior $\Gamma(\\alpha,\\beta)$:({},{})".format(params[i][0], params[i][1]))
legend()
show()
```

```python
condition_expect = []  # Здесь будут предсказанные значения
time = []  # Сюда записываю время с шагом t_0
quant = []  # Количество серверов, вышедших из строя, на данный момент

plt.figure(figsize=(18,7))
for i in range(t/t_0):
    time.append(t_0*(i+1))
    quant.append(shape(np.where(data <= time[-1]))[1])  # Нахожу это количество подсчетом
                                                        # числа записей времени, меньших
                                                        # текущего времени
    condition_expect.append(((t-time[-1])*lmd) + quant[-1])  # Предсказанное значение

plt.xlabel('Time', fontsize=15)
plt.ylabel('Number of Servers to Buy', fontsize=15)
plt.plot(time,condition_expect)
plt.grid()
plt.show()

plt.figure(figsize=(18,7))
plt.xlabel('Time', fontsize=15)
plt.ylabel('Number of Servers to Buy', fontsize=15)
plt.plot(time,condition_expect, label='Predicted number')
plt.plot(time,quant, color='g', label='Real number of damaged servers')
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, \
       mode="expand", borderaxespad=0.)
plt.show()
```
