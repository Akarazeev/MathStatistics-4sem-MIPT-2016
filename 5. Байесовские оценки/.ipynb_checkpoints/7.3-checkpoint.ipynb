```python
sz = 100
theta0 = 0
x = cauchy.rvs(size=sz)
means = np.array([x[:(i+1)].mean() for i in range(sz)])

plot(x)

figure(figsize=(12,5))
plot(abs(theta0-means), label='MLE')
legend()
show()
```
