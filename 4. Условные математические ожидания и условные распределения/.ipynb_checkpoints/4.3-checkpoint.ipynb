Воспользуемся линейностью условного матожа: $E(N_t|N_s) = E(N_t-N_s|N_s) + E(N_s|N_s)$. В условии сказано, что $(N_t-N_s) \sim Pois(\frac{t-s}{\lambda})$ и $(N_t-N_s)$ независимо c $N_s$. Следовательно, $E(N_t|N_s) = E(N_t-N_s) + N_s$ и $E(N_t|N_s) = \frac{t-s}{\lambda} + N_s$.

```python
lmd = 66
t_0 = 200
t = 70000

condition_expect = []
time = []
plt.figure(figsize=(10,10))
for i in range(t/t_0):
    time.append(t_0*(i+1))
    condition_expect.append(((t-time[-1])/lmd) + shape(np.where(data <= time[-1]))[1])
plt.xlabel('Time', fontsize=15)
plt.ylabel('Number of Servers to Buy', fontsize=15)
plt.plot(time,condition_expect)
plt.show()
```
