```python
%matplotlib inline
import numpy as np
import math as mt
import matplotlib
import matplotlib.pyplot as plt
from pylab import *
from scipy.stats import *
```

```python
def FindMaxArg(data):
    # Находим аргумент, при котором функция правдоподобия
    # принимает максимальное значение.
    # Для этого ищем значение a_min, при котором f 
    # принимает минимальное значение, где f - логарифм
    # от функции правдоподобия со знаком минус.
    a_min = 0.
    f_min = sum(1.+((data-a_min)**2.))
    step = 0.01
    it = -1000
    while it <= 1000:
        tmp = sum(1.+((data-it)**2.))
        if tmp<=f_min:
            a_min = it
            f_min = tmp
        it += step
    return a_min
```

```python
from numpy import genfromtxt
my_data = genfromtxt('Cauchy.csv', delimiter=',')
```

```python
ans = FindMaxArg(my_data[0:(len(my_data)/2)])
print ans

r = (-190,-150)

figure()
h = hist(my_data, range=r, bins=30)
x = np.linspace(r[0], r[1])
plot(x, max(h[0])*(cauchy.pdf(x,loc=ans)/(max(cauchy.pdf(x,loc=ans)))), linewidth=4)
show()
```
