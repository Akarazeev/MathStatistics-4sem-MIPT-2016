```python
# Импортируем всякие библиотеки
import numpy as np
import scipy.stats as sst
import matplotlib.pyplot as plt

%matplotlib inline

GRAPH_ALPHA = 0.5
GRAPH_LINEWIDTH = 0.1

SIZE = 1000
M = 50
P = 0.3
NUMBER_OF_SAMPLES = 500

firstDis = []
secondDis = []
infFisher = []

for i in range(1, SIZE):
    firstEstimations = []
    secondEstimations = []
    for j in  range(NUMBER_OF_SAMPLES):
        sample = sst.binom.rvs(M, P, size=i)  # Генерация выборки
        firstEstimations.append(np.mean(sample) / M)  
        secondEstimations.append(sample[0] / M)
    firstDis.append(np.var(firstEstimations))
    secondDis.append(np.var(secondEstimations))
    infFisher.append((P * (1 - P)) / (i * M))

grid = np.linspace(1, SIZE, SIZE - 1)

plt.figure(figsize=(18, 5))  # Инициализируем график и задаем его размеры
#plt.axis([0, SIZE, 0, 0.00002])
plt.plot(grid, firstDis, color='green', alpha=1, 
         label=r'$D_{\theta}(\hat{\theta})$ - estimation $\frac{\overline{X}}{m}$')
plt.plot(grid, secondDis, color='red', alpha=1, 
         label=r'$D_{\theta}(\hat{\theta})$ - estimation $\frac{X_1}{m}$')
plt.plot(grid, infFisher, color='blue', alpha=1, 
         label=r'$\frac{1}{D_{X}{\theta}}$')
plt.legend()  # Ставим легенду
plt.title('binomial')
plt.show()  # Показываем график

plt.figure(figsize=(18, 5))  # Инициализируем график и задаем его размеры
plt.axis([0, SIZE, 0, 0.00002])
plt.plot(grid, firstDis, color='green', alpha=1, 
         label=r'$D_{\theta}(\hat{\theta})$ - estimation $\frac{\overline{X}}{m}$')
plt.plot(grid, infFisher, color='blue', alpha=1, 
         label=r'$\frac{1}{D_{X}{\theta}}$')
plt.legend()  # Ставим легенду
plt.show()  # Показываем график
```

```python
# Импортируем всякие библиотеки
import numpy as np
import scipy.stats as sst
import matplotlib.pyplot as plt

%matplotlib inline

GRAPH_ALPHA = 0.5
GRAPH_LINEWIDTH = 0.1

SIZE = 1000
A = 3
SIGMA = 16
NUMBER_OF_SAMPLES = 50

firstDis = []
secondDis = []
infFisher = []

for i in range(1, SIZE):
    firstEstimations = []
    secondEstimations = []
    for j in  range(NUMBER_OF_SAMPLES):
        sample = sst.norm.rvs(A, SIGMA, size=i)  # Генерация выборки
        firstEstimations.append(np.mean((sample - A) ** 2))
        secondEstimations.append(np.median(sample))
    firstDis.append(np.var(firstEstimations))
    secondDis.append(np.var(secondEstimations))
    infFisher.append(1 / (i * 2 * (SIGMA ** 2)))

grid = np.linspace(1, SIZE, SIZE - 1)

plt.figure(figsize=(18, 5))  # Инициализируем график и задаем его размеры
#plt.axis([0, SIZE, 0, 0.00002])
plt.plot(grid, firstDis, color='green', alpha=1, 
         label=r'$D_{\theta}(\hat{\theta})$ - estimation $\frac{\overline{X}}{m}$')
plt.plot(grid, infFisher, color='blue', alpha=1, 
         label=r'$\frac{1}{D_{X}{\theta}}$')
plt.legend()  # Ставим легенду
plt.title(r'normal withot $\sigma$')
plt.show()  # Показываем график

plt.figure(figsize=(18, 5))  # Инициализируем график и задаем его размеры
plt.axis([0, SIZE, 0, 1000])
plt.plot(grid, firstDis, color='green', alpha=1, 
         label=r'$D_{\theta}(\hat{\theta})$ - estimation $\frac{\overline{X}}{m}$')
plt.plot(grid, infFisher, color='blue', alpha=1, 
         label=r'$\frac{1}{D_{X}{\theta}}$')
plt.legend()  # Ставим легенду
plt.show()  # Показываем график
```

```python
# Импортируем всякие библиотеки
import numpy as np
import scipy.stats as sst
import matplotlib.pyplot as plt

%matplotlib inline

GRAPH_ALPHA = 0.5
GRAPH_LINEWIDTH = 0.1

SIZE = 1000
A = 3
SIGMA = 16
NUMBER_OF_SAMPLES = 50

firstDis = []
secondDis = []
infFisher = []

for i in range(1, SIZE):
    firstEstimations = []
    secondEstimations = []
    for j in  range(NUMBER_OF_SAMPLES):
        sample = sst.norm.rvs(A, SIGMA, size=i)  # Генерация выборки
        firstEstimations.append(np.mean(sample))
        secondEstimations.append(np.median(sample))
    firstDis.append(np.var(firstEstimations))
    secondDis.append(np.var(secondEstimations))
    infFisher.append((SIGMA ** 2) / i)

grid = np.linspace(1, SIZE, SIZE - 1)

plt.figure(figsize=(18, 5))  # Инициализируем график и задаем его размеры
#plt.axis([0, SIZE, 0, 0.00002])
plt.plot(grid, firstDis, color='green', alpha=1, 
         label=r'$D_{\theta}(\hat{\theta})$ - estimation $\frac{\overline{X}}{m}$')
plt.plot(grid, infFisher, color='blue', alpha=1, 
         label=r'$\frac{1}{D_{X}{\theta}}$')
plt.legend()  # Ставим легенду
plt.title(r'normal without a')
plt.show()  # Показываем график

plt.figure(figsize=(18, 5))  # Инициализируем график и задаем его размеры
plt.axis([0, SIZE, 0, 1])
plt.plot(grid, firstDis, color='green', alpha=1, 
         label=r'$D_{\theta}(\hat{\theta})$ - estimation $\frac{\overline{X}}{m}$')
plt.plot(grid, infFisher, color='blue', alpha=1, 
         label=r'$\frac{1}{D_{X}{\theta}}$')
plt.legend()  # Ставим легенду
plt.show()  # Показываем график
```
