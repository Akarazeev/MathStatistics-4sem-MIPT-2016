```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from pylab import *
```

```python
def Uniform(theta, N):
    s = np.random.uniform(0,theta,N)
    
    a = np.array([(2*sum(s[0:n]))/n for n in range(1,N)])
    b = np.array([(np.partition(np.asarray(s),n)[n]/2)+((sum(s[0:n]))/n) for n in range(1,N)])
    c = np.array([(n+1)*min(s) for n in range(1,N)])
    d = np.array([min(s)+np.partition(np.asarray(s),n)[n] for n in range(1,N)])
    e = np.array([((n+1)/n)*np.partition(np.asarray(s),n)[n] for n in range(1,N)])
    
#     all = np.array([a,b,c,d,e])
#     for i in range(len(all)):
#         print np.mean(all[i])
    
    figure()
    plot(abs(a-theta), 'r')
    plot(abs(b-theta), 'g')
    plot(abs(c-theta), 'b')
    plot(abs(d-theta), 'y')
    plot(abs(e-theta), 'pink')
    show()
```

```python
Uniform(10, 10000)
```

```python
Uniform(2, 10000)
```
