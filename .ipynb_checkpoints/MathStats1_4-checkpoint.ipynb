```python
import numpy as np
import scipy.stats as sst
import matplotlib.pyplot as plt
import math

%matplotlib inline

SIZE = 10000

CURRENT_VAR = 0.1311


GRAPH_LINEWIDTH = 0.1
GRAPH_ALPHA = 0.5

COLORS = ['blue', 'green', 'cyan', 'magenta', 'yellow', 'black']

sample = sst.norm.rvs(size=SIZE, loc=1, scale=3)

grid = np.linspace(-10, 10, SIZE)  # задаем сетку для построения графика плотности

needN = [10, 20, 50, 100, 1000, SIZE]
currColor = 0

for index in needN:   
    currSample = sorted(sample[:index])
    grid = np.linspace(-10, 10, SIZE)
    empire = np.zeros(SIZE);
    jumps = [];
    currIndexOnSample = 0

    for x, elem in enumerate(grid):
        if x != 0:
            empire[x] = empire[x - 1]
        while currIndexOnSample < index and currSample[currIndexOnSample] < elem:
            empire[x] += 1
            currIndexOnSample += 1
            jumps.append(elem)
    empire /= index

    plt.figure(figsize=(18, 5))
    plt.axis([-10, 10, -0.1, 1.1])
    plt.scatter(jumps, np.zeros(len(jumps)) - 0.02, alpha=0.2, label='sample')  # label - описание в легенде
    plt.plot(grid, sst.norm.cdf(grid, loc=1, scale=3), color='red', label='density')  # color - цвет графика

    plt.plot(grid, empire, color=COLORS[currColor], label='n = ' + str(index), alpha = 1)  # color - цвет графика
    currColor += 1

    plt.title(r'n = ' + str(index)) 
    plt.legend()  # добавляет легенду
    plt.show()
```

```python
import numpy as np
import scipy.stats as sst
import matplotlib.pyplot as plt
import math

%matplotlib inline

SIZE = 1000

CURRENT_VAR = 0.1311


GRAPH_LINEWIDTH = 0.1
GRAPH_ALPHA = 0.5

COLORS = ['blue', 'green', 'cyan', 'magenta', 'yellow', 'black']

sample = sst.norm.rvs(size=SIZE, loc=1, scale=3)


suprenums = []
grid = np.linspace(0, SIZE, SIZE)
for n in range(SIZE):
    currN = n
    if n == 0:
        suprenums.append(0)
        continue
    currSample = sorted(sample[:currN])
    empire = np.zeros(SIZE);
    currIndexOnSample = 0

    for x, elem in enumerate(grid):
        if x != 0:
            empire[x] = empire[x - 1]
        while currIndexOnSample < currN and currSample[currIndexOnSample] < elem:
            empire[x] += 1
            currIndexOnSample += 1
    empire /= currN
    
    currD = 0
    for x in range(SIZE):
        currD = max(currD, abs(empire[x] - sst.norm.cdf(grid[x], loc = 1, scale=3)))
    suprenums.append(currD)
    
    
plt.figure(figsize=(18, 7))
plt.axis([0, SIZE, 0, 0.2])
plt.plot(grid, suprenums, color='blue', label=r'$D_n')  # color - цвет графика
plt.title(r'$D_n$') 
plt.show()
```
