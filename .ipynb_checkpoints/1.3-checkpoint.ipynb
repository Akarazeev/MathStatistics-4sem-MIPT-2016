```python
%matplotlib inline
import numpy as np
import math as mt
import matplotlib
import matplotlib.pyplot as plt
from pylab import *
from scipy.stats import *
```

```python
N = 10000

a = 5
s_par = pareto.rvs(a, size=N)

figure()
right_lim = 4
xlim((0.9, right_lim))
x_par = np.linspace(1, right_lim)
plot(x, pareto.pdf(x_par, a), 'r-', linewidth=1.0, label='Pareto PDF')
scatter(s_par, np.zeros(N), s=60, color='g', alpha=0.1, label='Pareto variates')
legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)
show()

par_var_array = np.array([var(s_par[0:n]) for n in range(1,N)])
figure()
plot(abs(par_var_array - pareto.var(a)), 'r')
title('$| s^2 - variance |$')
legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)
show()
```

```python
N = 10000

s_cau = cauchy.rvs(size=N)
lim = 100

figure()
x_cau = np.linspace(-lim, lim)
xlim((-lim, lim))
ylim()
plot(x_cau, cauchy.pdf(x_cau), 'r-', linewidth=0.8, label='Cauchy PDF')
scatter(s_cau, np.zeros(N), s=60, color='g', alpha=0.1, label='Cauchy variates')
legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)
show()

cau_var_array = np.array([var(s_cau[0:n]) for n in range(1,N)])
figure()
plot(cau_var_array, 'r')
title('$| s^2 - variance |$')
show()
```
