```python
%matplotlib inline
import numpy as np
import math as mt
import matplotlib
import matplotlib.pyplot as plt
from pylab import *
from scipy.stats import *
```

```python
def CumFun(xs, n):
    step = np.linspace(xs[0:n].min(), xs[0:n].max(), n)
    res = np.zeros(n)
    for i in range(n):
        res[i] = sum(k<=step[i] for k in xs[0:n])/float(n)
    return (step, res)

N = 10000
s_nor = norm.rvs(size=N)

ns = [10,25,50,100,1000,N]

for i in ns:
    figure()
    x = np.linspace(-4,4)
    plot(x, norm.cdf(x), 'b')
    
    x, y = CumFun(s_nor,i)
    plot(x, y, 'r')
    
    scatter(s_nor[0:i], np.zeros(i), s=60, color='g', alpha=0.2)
    
    show()
```

```python
s_nor = norm.rvs(size=N)
dev = np.zeros(N)

for i in range(1,N+1,1000):
    print i
    x, y = CumFun(s_nor,i)
    dev[i-1] = max(abs(norm.cdf(x)-y))

figure()
plot(dev)
show()
```

```python
def CumFunN(xs, n):
    step = np.linspace(xs[0:n].min(), xs[0:n].max(), 100)
    res = np.zeros(n)
    for i in range(n):
        res[i] = sum(k<=step[i/100] for k in xs[0:n])/float(n)
    return (step, res)

s_norr = norm.rvs(size=N)
devv = np.zeros(N)

for i in range(1,N+1,1000):
    print i
    x, y = CumFunN(s_norr,i)
    devv[i-1] = max(abs(norm.cdf(x)-y))

figure()
plot(devv)
show()
```
